## Write a function clean_gender(val) that standardises the value and Apply it to create a new column  Gender_Clean

```python
def clean_gender(val):
  if pd.isna(val):
    return np.nan

  val = str(val).strip().lower()

  if val in ['male', 'm', '1']:
    return 'Male'
  elif val in ['female','f','0']:
    return 'Female'
  else:
    return np.nan

df['Gender_Clean'] = df['Gender'].apply(clean_gender)
```

# FIXING MIXED-FORMAT COLUMNS

a) SALARY column contains values like "45000 rupees", "45k", "not disclosed", empty strings.  
- Convert everything to a numeric float (NaN where impossible).
- Store result in  Salary_Clean.

```python
df['Salary'].unique()
```

```python
def clean_salary(val):
  if pd.isna(val):
    return np.nan

  val = str(val).strip().lower()

  if val in ['', 'not disclose']:
    return np.nan
  val = val.replace('rupees', '').replace('k', '000').strip()

  try:
    return float(val)
  except:
    return np.nan

df['Salary_Clean'] = df['Salary'].apply(clean_salary)
```

```python
df['Salary_Clean'].unique()
```

b) AGE column has values like "35 years", 999, -5.
- Convert to numeric, replace impossible ages (< 18 or > 80) with NaN.
- Store result in  Age_Clean.

```python
df['Age'].unique()
```

```python
def clean_age(val):
  if pd.isna(val):
    return np.nan

  val = str(val).strip().lower()
  val = val.replace('years','').strip()

  for a in val:
    if a < '18' or a > '80':
      return np.nan

  try:
    return int(val)
  except:
    return np.nan

df['Age_Clean'] = df['Age'].apply(clean_age)
```

```python
df['Age_Clean'].unique()
```

c) PERFORMANCE_SCORE column has values like "78%", 150, -1.
- Convert to numeric and cap valid range to 0–100 (replace outliers with NaN).
- Store result in  Score_Clean.

```python
df['Performance_Score'].unique()
```

```python
def clean_score(val):
  if pd.isna(val):
    return np.nan

  val = str(val).strip().lower()
  val = val.replace('%','').strip()


  try:
    score = float(val)
    if 0 <= score <=100:
      return score
    else:
      return np.nan
  except:
    return np.nan

df['Score_Clean'] = df['Performance_Score'].apply(clean_score)
```

```python
df['Score_Clean'].unique()
```

d) EXPERIENCE_YEARS has values like "5 yrs", "fresher".
- Convert "fresher" → 0, strip " yrs", convert to numeric.
- Store result in  Experience_Clean.

```python
df['Experience_Years'].unique()
```

```python
def clean_experience(val):
  if pd.isna(val):
    return np.nan

  val = str(val).strip().lower()
  val = val.replace('yrs','').replace('fresher','0').strip()


  try:
    return float(val)
  except:
    return np.nan

df['Experience_Clean'] = df['Experience_Years'].apply(clean_experience)
```

## ORDINAL ENCODING — Education

Education has a clear ORDER: High School < Diploma < Bachelor's < Master's < PhD

a) Define the ordered category list

b) Use sklearn OrdinalEncoder with the categories defined in (a)

c) Store result in  Education_Encoded  (integer 1–5)

d) Print a cross-tab of Education vs Education_Encoded to verify correctness

e) Why would Label Encoding be WRONG here?  (comment answer)

Answer:
Label Encoding is not appropriate because it assigns arbitrary numeric values to categories without preserving their logical order.

In the Education column, there is a clear hierarchy:
High School < Diploma < Bachelor's < Master's < PhD

Label Encoding might assign:
High School = 3, PhD = 1 (random order)

This misleads the model, as it does not reflect the true progression.
Ordinal Encoding is correct because it preserves the natural ranking

```python
education_order = ['High School', 'Diploma', 'Bachelor\'s', 'Master\'s', 'PhD']
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(categories=[education_order], dtype=int)
df['Education_Encoded'] = encoder.fit_transform(df[['Education']])
print(df[['Education', 'Education_Encoded']])
crosstab = pd.crosstab(df['Education'], df['Education_Encoded'])
print(crosstab)
```

## OUTLIER DETECTION — IQR METHOD on Salary

a) Calculate Q1, Q3, and IQR for Salary_Clean

b) Define lower and upper bounds: Q1 − 1.5×IQR  and  Q3 + 1.5×IQR

c) Print count of outliers below lower bound and above upper bound

d) Print the actual outlier salary values

e) TREAT outliers using CAPPING (Winsorization):
- replace values below lower with lower, above upper with upper
- Store result in  Salary_Capped

f) Verify: print min and max of Salary_Capped

```python
Q1 = df['Salary_Clean'].quantile(0.25)
Q3 = df['Salary_Clean'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_below = (df['Salary_Clean'] < lower_bound).sum()
outliers_above = (df['Salary_Clean'] > upper_bound).sum()
print(f"Below: {outliers_below}, Above: {outliers_above}")

print("Outliers:", df[(df['Salary_Clean'] < lower_bound) |
                       (df['Salary_Clean'] > upper_bound)]['Salary_Clean'].values)

df['Salary_Capped'] = df['Salary_Clean'].copy()
df.loc[df['Salary_Capped'] < lower_bound, 'Salary_Capped'] = lower_bound
df.loc[df['Salary_Capped'] > upper_bound, 'Salary_Capped'] = upper_bound

print(f"Min: {df['Salary_Capped'].min():,.2f}")
print(f"Max: {df['Salary_Capped'].max():,.2f}")
```

## OUTLIER DETECTION — Z-SCORE METHOD on Score_Clean

a) Calculate Z-score for every value in Score_Clean
- Z = (x − mean) / std

b) Flag rows where |Z| > 3 as outliers

c) Print count of flagged outliers

d) Remove (drop) outlier rows from df for Score_Clean

e) Print shape before and after removal

```python
mean = df['Score_Clean'].mean()
std = df['Score_Clean'].std()
df['Z_Score'] = (df['Score_Clean'] - mean) / std

df['Is_Outlier'] = np.abs(df['Z_Score']) > 3

outlier_count = df['Is_Outlier'].sum()
print(f"Outliers detected: {outlier_count}")
print(df[df['Is_Outlier']][['Score_Clean', 'Z_Score']])

df_cleaned = df[~df['Is_Outlier']].copy()

print(f"\nBEFORE: {df.shape}")
print(f"AFTER:  {df_cleaned.shape}")
print(f"Removed: {len(df) - len(df_cleaned)} rows")
```

## FEATURE CREATION — Date-based Features

Join_Date column has mixed formats (YYYY-MM-DD, DD/MM/YYYY, "invalid", etc.)

a) Parse Join_Date to datetime using pd.to_datetime with errors='coerce'

b) Fill NaT values with the median date

c) Create new column  Tenure_Years  = (today's date − join date) / 365
- Round to 1 decimal

d) Create new column  Join_Year   (integer year)

e) Create new column  Join_Quarter  (1, 2, 3, or 4)

```python
df["Join_Date"] = pd.to_datetime(df["Join_Date"], errors="coerce")

median_date = df["Join_Date"].median()
df["Join_Date"].fillna(median_date, inplace=True)

today = pd.Timestamp.today()

df["Tenure_Years"] = ((today - df["Join_Date"]).dt.days / 365).round(1)
df["Join_Year"] = df["Join_Date"].dt.year
df["Join_Quarter"] = df["Join_Date"].dt.quarter
```

## FEATURE CREATION — Derived Numeric Features

a) Create  Salary_Per_Year_Exp  = Salary_Capped / (Experience_Clean + 1)
- (+1 avoids division by zero for freshers)
- Round to 2 decimals.

b) Create  Score_Salary_Ratio  = Score_Clean / Salary_Capped * 1000
- (performance per ₹1000 earned — useful for ML ranking tasks)
- Round to 4 decimals.

c) Create  Is_Senior  = 1 if Age_Clean >= 40 else 0
- (binary flag for senior employees)

d) Create  Exp_Age_Ratio  = Experience_Clean / Age_Clean
- (what fraction of their life has been working experience?)
- Round to 3 decimals.

e) Print describe() for all 4 new features

```python
df["Salary_Per_Year_Exp"] = (df["Salary_Capped"] /
                             (df["Experience_Clean"] + 1)).round(2)

df["Score_Salary_Ratio"] = (
    df["Score_Clean"] / df["Salary_Capped"] * 1000
).round(4)

df["Is_Senior"] = np.where(df["Age_Clean"] >= 40, 1, 0)

df["Exp_Age_Ratio"] = (
    df["Experience_Clean"] / df["Age_Clean"]
).round(3)

print(df[["Salary_Per_Year_Exp",
          "Score_Salary_Ratio",
          "Is_Senior",
          "Exp_Age_Ratio"]].describe())
```

## BONUS — FREQUENCY ENCODING on City

Frequency encoding replaces each category with how many times it appears.
This is useful when many unique categories exist.

a) Starting from df_raw, clean the City column (strip + title-case)

b) Compute frequency count of each city

c) Create column  City_Freq  by mapping each city to its frequency count

d) Normalise City_Freq to 0–1 using Min-Max scaling and store as  City_Freq_Norm

e) What happens if two different cities have the same frequency?
Is that a problem? (comment your answer)

Answer:
If two different cities have the same frequency:

- They will get the same encoded value

This leads to loss of information, because:
- Different categories become indistinguishable

However:

- This is generally acceptable
- Because frequency encoding focuses on importance/popularity, not identity

```python
df3 = df_raw.copy()

df3["City"] = df3["City"].str.strip().str.title()

freq = df3["City"].value_counts()

df3["City_Freq"] = df3["City"].map(freq)

# Min-Max normalization
df3["City_Freq_Norm"] = (
    (df3["City_Freq"] - df3["City_Freq"].min()) /
    (df3["City_Freq"].max() - df3["City_Freq"].min())
)

print(df3[["City","City_Freq","City_Freq_Norm"]].head())
```
